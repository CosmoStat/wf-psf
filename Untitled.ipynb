## Need to compute unweighted centroid

```python
def compute_centroid(height, width, image):
    #  height = image_x.shape[0]
    #  width = image_y.shape[0]

    # Create coordinate grids
    x, y = np.meshgrid(height, width)

    # Calculate the centroid coordinates
    total_intensity = np.sum(image)
    cx = np.sum(x * image) / total_intensity
    cy = np.sum(y * image) / total_intensity

    return cx, cy


def scale_to_range(input_array, old_range, new_range):
    # Scale to [0,1]
    input_array = (input_array - old_range[0]) / (old_range[1] - old_range[0])
    # Scale to new_range
    input_array = input_array * (new_range[1] - new_range[0]) + new_range[0]
    return input_array


def rescale_2d_image(image):
    # Scale positions to a new range
    x_lims = [0, 1e3]
    y_lims = [0, 1e3]
    # X coordinate
    x_old_range = [np.min(image["positions"][:, 0]), np.max(image["positions"][:, 0])]
    image_x = scale_to_range(image["positions"][:, 0], x_old_range, x_lims)

    # Y coordinate
    y_old_range = [
        np.min(image["positions"][:, 1]),
        np.max(wf_dataset["positions"][:, 1]),
    ]
    image_y = scale_to_range(image["positions"][:, 1], y_old_range, y_lims)
    image["positions"][:, 0] = vis_x
    image["positions"][:, 1] = vis_y

    return image
```

```python
import numpy as np


def split_data(data_dict, split_ratio=0.8, random_seed=None):
    """
    Split a dictionary into train and test datasets.

    Parameters:
    - data_dict: Input dictionary containing data.
    - split_ratio: Ratio of data to be used for training (default is 0.8).
    - random_seed: Seed for reproducibility (default is None).

    Returns:
    - train_data: Dictionary containing training data.
    - test_data: Dictionary containing test data.
    """

    # Set random seed for reproducibility
    if random_seed is not None:
        np.random.seed(random_seed)

    # Extract keys and values from the dictionary
    keys = list(data_dict.keys())
    values = list(data_dict.values())

    # Get the number of samples
    num_samples = len(values[0])

    # Shuffle indices to randomize the data
    indices = np.arange(num_samples)
    np.random.shuffle(indices)

    # Calculate the number of samples for training
    num_train = int(split_ratio * num_samples)

    # Split indices into training and test sets
    train_indices = indices[:num_train]
    test_indices = indices[num_train:]

    # Create dictionaries for training and test data
    train_data = {
        key: np.array([values[i] for i in train_indices])
        for key, values in data_dict.items()
    }
    test_data = {
        key: np.array([values[i] for i in test_indices])
        for key, values in data_dict.items()
    }

    return train_data, test_data


# Example usage
data_dict = {
    "feature1": [1, 2, 3, 4, 5],
    "feature2": [6, 7, 8, 9, 10],
    "target": [0, 1, 0, 1, 0],
}

train_data, test_data = split_data(data_dict, split_ratio=0.8, random_seed=42)

print("Train data:")
print(train_data)

print("\nTest data:")
print(test_data)
```
