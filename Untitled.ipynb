```python
def simple_star_and_mask():
    """Fixture for an image with multiple non-zero pixels for centroid calculation."""
    num_images = 5  # Change this to test with multiple images
    image = np.zeros(
        (num_images, 5, 5)
    )  # Create a 3D array (5x5 image for each "image")

    # Place non-zero values in multiple pixels
    image[:, 2, 2] = 10  # Star at the center
    image[:, 2, 3] = 10  # Adjacent pixel
    image[:, 3, 2] = 10  # Adjacent pixel
    image[:, 3, 3] = 10  # Adjacent pixel forming a symmetric pattern

    mask = np.zeros_like(image)
    mask[:, 3, 2] = 1
    mask[:, 3, 3] = 1

    return image, mask
```

```python
print(zernike_contribution_list[0].shape)
print(zernike_contribution_list[1].shape)
```

```python
if len(zernike_contribution_list) == 1:
    zernike_contribution = zernike_contribution_list[0]
else:
    # Get max zk order
    max_zk_order = np.max(
        np.array(
            [zk_contribution.shape[1] for zk_contribution in zernike_contribution_list]
        )
    )

    zernike_contribution = np.zeros(
        (zernike_contribution_list[0].shape[0], max_zk_order)
    )

    # Pad arrays to get the same length and add the final contribution
    for it in range(len(zernike_contribution_list)):
        current_zk_order = zernike_contribution_list[it].shape[1]
        current_zernike_contribution = np.pad(
            zernike_contribution_list[it],
            pad_width=[(0, 0), (0, int(max_zk_order - current_zk_order))],
            mode="constant",
            constant_values=0,
        )

        zernike_contribution += current_zernike_contribution
```

```python
old_centroid = old_cc(model_params, data, batch_size)
new_centroid = new_cc(model_params, zks_inputs.centroid_dataset, batch_size)


old_ccd = old_ccd_mas(model_params, data)
new_ccd = new_ccd_mas(model_params, positions)

import numpy as np
import pandas as pd


def report_delta(a, b, name):
    diff = a - b
    if np.allclose(a, b):
        print(f"{name}: identical")
        return
    idx = np.unravel_index(np.abs(diff).argmax(), diff.shape)
    print(f"{name}: max Δ={diff[idx]} at {idx} " f"(old={a[idx]}, new={b[idx]})")


report_delta(old_centroid, new_centroid, "centroid")
report_delta(old_ccd, new_ccd, "ccd")

# Optional: put them in a DataFrame to eyeball a few rows
pd.set_option("display.precision", 8)
display(
    pd.DataFrame(
        {
            "old_tip": old_centroid[:, 0],
            "new_tip": new_centroid[:, 0],
            "old_tilt": old_centroid[:, 1],
            "new_tilt": new_centroid[:, 1],
        }
    )
)
```

```python
from wf_psf.data.data_zernike_utils import (
    ZernikeInputsFactory,
    assemble_zernike_contributions,
)


def _assemble_zernike_contributions(data, run_type, model_params, prior=None):
    zks_inputs = ZernikeInputsFactory.build(
        data=data,
        run_type=run_type,
        model_params=model_params,
        prior=prior,
    )
    return assemble_zernike_contributions(
        model_params=model_params,
        zernike_prior=zks_inputs.zernike_prior,
        centroid_dataset=zks_inputs.centroid_dataset,
        positions=zks_inputs.misalignment_positions,
        batch_size=16,
    )
```
